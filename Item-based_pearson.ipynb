```python
X_train = X_train.pivot_table(index = 'courseId', columns = 'userId', values = 'rating').fillna(0)
df1 = X_train.copy()
```

```python
def get_coursename(courseid):
  return (courses.loc[(courses.courseId == courseid), 'name'].iloc[0])
```

```python
def recommend_courses(user, df):

  recommended_courses = []
  recommendations = []

  for m in df[df[user] == 0].index.tolist():

    index_df = df.index.tolist().index(m)
    predicted_rating = df1.iloc[index_df, df1.columns.tolist().index(user)]
    recommended_courses.append((m, predicted_rating))

  sorted_rm = sorted(recommended_courses, key = lambda x:x[1], reverse = True)
  
  for recommended_course in sorted_rm:
    recommendations.append(((recommended_course[0], get_coursename(recommended_course[0]), recommended_course[1])))
  return recommendations
```

```python
from sklearn.neighbors import NearestNeighbors
from scipy.spatial.distance import correlation
def get_recommendation(user, num_neighbors, df):
# find the nearest neighbors using NearestNeighbors(n_neighbors=3)
  number_neighbors = num_neighbors
  knn = NearestNeighbors(metric = 'correlation')
  knn.fit(df.values)
  distances, indices = knn.kneighbors(df.values, n_neighbors = number_neighbors)

  #indices shows the nearest courses to each course. 
  #Each row corresponds to the row in the df. The first element in a row is the most similar (nearest) course. 
  #distances shows the distance between courses. Each number in this matrix corresponds to the number in the indices matrix.

  # convert user_name to user_index
  user_index = df.columns.tolist().index(user)

  # t: course_id, m: the row number of t in df
  for m,t in list(enumerate(df.index)):
    
    # find courses without ratings by user
    if df.iloc[m, user_index] == 0:
      sim_courses = indices[m].tolist()
      course_distances = distances[m].tolist()
      
      if m in sim_courses:
        id_course = sim_courses.index(m)
        sim_courses.remove(m)
        course_distances.pop(id_course) 

      else:
        sim_courses = sim_courses[:number_neighbors-1]
        course_distances = course_distances[:number_neighbors-1]
          
      # course_similarty = 1 - course_distance    
      course_similarity = [1-x for x in course_distances]
      course_similarity_copy = course_similarity.copy()
      nominator = 0

      # for each similar course
      for s in range(0, len(course_similarity)):
        
        # check if the rating of a similar course is zero
        if df.iloc[sim_courses[s], user_index] == 0:

          # if the rating is zero, ignore the rating and the similarity in calculating the predicted rating
          if len(course_similarity_copy) == (number_neighbors - 1):
            course_similarity_copy.pop(s)
            
          else:
            course_similarity_copy.pop(s-(len(course_similarity)-len(course_similarity_copy)))

        # if the rating is not zero, use the rating and similarity in the calculation
        else:
          nominator = nominator + course_similarity[s]*df.iloc[sim_courses[s],user_index]

      # check if the number of the ratings with non-zero is positive
      if len(course_similarity_copy) > 0:
        
        # check if the sum of the ratings of the similar courses is positive.
        if sum(course_similarity_copy) > 0:
          predicted_r = nominator/sum(course_similarity_copy)
        # Even if there are some courses for which the ratings are positive, some courses have zero similarity even though they are selected as similar courses.
        # in this case, the predicted rating becomes zero as well  
        else:
          predicted_r = 0

      # if all the ratings of the similar courses are zero, then predicted rating should be zero
      else:
        predicted_r = 0

    # place the predicted rating into the copy of the original dataset
      df1.iloc[m,user_index] = predicted_r
  return recommend_courses(user, df)
```

```python
def get_recommendation_4test(test):

  X_test_pred = []
  X_test_userid = X_test['userId'].tolist()   
  X_test_courseid = X_test['courseId'].tolist()
 
  print("------Dự đoán cho tập test với {} điểm dữ liệu -----------".format(len(X_test_userid)))
  
  for i in range(len(X_test_userid)):        
    list_R = get_recommendation(X_test_userid[i], 10, X_train) 
    check = 0
    for j in list_R:
      if(X_test_courseid[i] == j[0]):     
        X_test_pred.append(j[2])
        check = 1
    if(check == 0):
      X_test_pred.append(0)
    print('----- Đang dự đoán cho dòng thứ {}'.format(i), 'là {}'.format(X_test_pred[i])) 
  return X_test_pred
```

```python
import time
start = time.time()

X_test_pred_rating = get_recommendation_4test(X_test)

end = time.time()
print('Tổng thời gian dự đoán: ', end - start)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math

mse = mean_squared_error(X_test['rating'].tolist(), X_test_pred_rating)
print('----- MSE: ', mse)

rmse = math.sqrt(mse)
print('----- RMSE: ', rmse)

mae = mean_absolute_error(X_test['rating'].tolist(), X_test_pred_rating)
print('----- MAE: ',mae)

max = X_test['rating'].max()
min = X_test['rating'].min()
nmae = mae/(max-min)
print('----- NMAE: ', nmae)
```
