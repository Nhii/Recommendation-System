# Recommendation

```python
#lấy ra rating của một user cho 1 khoá học
def get_rating(userid, courseid, ratings):
  return (ratings.loc[(ratings.userId == userid) & (ratings.courseId == courseid),'rating'].iloc[0])
#danh sách các khoá học đã rating bởi người dùng
def get_course_ids(userid, ratings):
    return (ratings.loc[(ratings.userId == userid),'courseId'].tolist())
#lấy ra tên của một khoá học
def get_course_title(courseid):
    return (courses.loc[(courses.courseId == courseid),'name'].iloc[0])
```

```python
from numpy import sqrt

def pearson_correlation_score(user1, user2, ratings):
  both_learn_count = [] # danh sách các khoá học được học chung bởi user1 và user2
  list_course_user1 = ratings.loc[ratings.userId == user1, 'courseId'].to_list()
  list_course_user2 = ratings.loc[ratings.userId == user2, 'courseId'].to_list()

  # danh sách các khoá học được học chung bởi user1 và user2
  for element in list_course_user1:
    if element in list_course_user2:
       both_learn_count.append(element)

  if(len(both_learn_count) == 0):
    return 0
  
  rating_sum_1 = sum([get_rating(user1, i, ratings) for i in both_learn_count])
  avg_rating_sum_1 = rating_sum_1/len(both_learn_count) # rating trung bình user1

  rating_sum_2 = sum([get_rating(user2, i, ratings) for i in both_learn_count])
  avg_rating_sum_2 = rating_sum_2/len(both_learn_count)# rating trung bình user2

  tu = sum([(get_rating(user1, i, ratings) - avg_rating_sum_1)*(get_rating(user2, i, ratings) - avg_rating_sum_2)  for i in both_learn_count])
  mau = sqrt(sum([pow((get_rating(user1, i, ratings) - avg_rating_sum_1),2) for i in both_learn_count]))*sqrt(sum([pow((get_rating(user2, i, ratings) - avg_rating_sum_2),2) for i in both_learn_count]))

  if(mau == 0):
    return 0
  
  return tu/mau
```

```python
def get_recommendation(userid, ratings):
    user_ids = ratings.userId.unique().tolist()
    total = {}
    similariy_sum = {}

    # Iterating over subset of user ids.
    for user in user_ids:
        
        # not comparing the user to itself (obviously!)
        if user == userid:
            continue
        
        # Getting similarity score between the users.
        score = pearson_correlation_score(userid, user, ratings)
        
        # not considering users having zero or less similarity score.
        if score <= 0:
            continue

        for courseid in get_course_ids(user, ratings):
            # Only considering not learned/rated movies
          if courseid not in get_course_ids(userid, ratings):
            total[courseid] = 0
            total[courseid] = get_rating(user, courseid, ratings) * score
            similariy_sum[courseid] = 0
            similariy_sum[courseid] += score

    # Normalizing ratings
    ranking = [(tot/similariy_sum[courseid], courseid) for courseid, tot in total.items()]
    ranking.sort() 
    ranking.reverse() 
    
    recommendations = [(get_course_title(courseid), courseid, score) for score, courseid in ranking]
    return recommendations
```

```python
def get_recommendation_4test(test):

  X_test_pred = []
  X_test_userid = X_test['userId'].tolist()   
  X_test_courseid = X_test['courseId'].tolist()
 
  print("------Dự đoán cho tập test với {} điểm dữ liệu -----------".format(len(X_test_userid)))
  
  for i in range(len(X_test_userid)):       
    list_R = get_recommendation(X_test_userid[i], X_train) 
    check = 0
    for j in list_R:
      if(X_test_courseid[i] == j[1]):    
        X_test_pred.append(j[2])
        check = 1
    if(check == 0):
      X_test_pred.append(0)
    print('----- Đang dự đoán cho dòng thứ {}'.format(i), 'là {}'.format(X_test_pred[i])) 
  return X_test_pred
```

```python
import time
start = time.time()

X_test_pred_rating = get_recommendation_4test(X_test)

end = time.time()
print('Tổng thời gian dự đoán: ', end - start)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math

mse = mean_squared_error(X_test['rating'].tolist(), X_test_pred_rating)
print('----- MSE: ', mse)

rmse = math.sqrt(mse)
print('----- RMSE: ', rmse)

mae = mean_absolute_error(X_test['rating'].tolist(), X_test_pred_rating)
print('----- MAE: ',mae)

max = X_test['rating'].max()
min = X_test['rating'].min()
nmae = mae/(max-min)
print('----- NMAE: ', nmae)
```
